```python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def insertion_sort(arr):
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are greater than key,
        # to one position ahead of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Example usage (you can test with partially sorted arrays)
# Example of a partially sorted array:
partially_sorted_array = [2, 5, 3, 8, 6, 9, 1]
# Example of a nearly sorted array:
nearly_sorted_array = [1, 3, 2, 5, 4, 6, 8, 7, 9]


print("Bubble Sort on partially sorted array:", bubble_sort(partially_sorted_array.copy()))
print("Insertion Sort on partially sorted array:", insertion_sort(partially_sorted_array.copy()))

print("\nBubble Sort on nearly sorted array:", bubble_sort(nearly_sorted_array.copy()))
print("Insertion Sort on nearly sorted array:", insertion_sort(nearly_sorted_array.copy()))
```

**Why Insertion Sort is more efficient for partially sorted data?**

Insertion Sort is more efficient for partially sorted arrays because of its adaptive nature. When an array is partially sorted, Insertion Sort requires fewer comparisons and shifts as it iterates through the elements. This is because elements are often already close to their correct positions, so the inner while loop that shifts elements terminates quickly. In contrast, Bubble Sort always performs a fixed number of comparisons and swaps (O(n^2) in the worst and average cases) regardless of the initial sortedness of the array, unless an optimization for early termination is added. Even with that optimization, Insertion Sort still tends to outperform Bubble Sort on nearly sorted data due to the way it handles insertions.

```python
def linear_search(arr, target):
    """
    Performs a linear search to find the target element in the array.

    Args:
        arr: The list or array to search within.
        target: The element to search for.

    Returns:
        The index of the target element if found, otherwise -1.

    Performance Notes:
        - Time Complexity: O(n) in the worst and average cases, where n is the number of elements in the array.
          In the best case (target is the first element), it is O(1).
        - It checks each element sequentially, making it suitable for small or unsorted arrays.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    """
    Performs a binary search to find the target element in a sorted array.

    Args:
        arr: The sorted list or array to search within.
        target: The element to search for.

    Returns:
        The index of the target element if found, otherwise -1.

    Performance Notes:
        - Time Complexity: O(log n) in the worst, average, and best cases, where n is the number of elements in the array.
        - It works by repeatedly dividing the search interval in half.
        - Requires the input array to be sorted for correct operation.
        - Significantly faster than linear search for large, sorted arrays.
    """
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Example usage
my_list_unsorted = [1, 5, 2, 8, 3, 9, 4, 7, 6]
my_list_sorted = sorted(my_list_unsorted) # Binary search requires a sorted list

target_element_found = 5
target_element_not_found = 10

print("Testing Linear Search:")
linear_result_found = linear_search(my_list_unsorted, target_element_found)
print(f"Linear Search (unsorted): Element {target_element_found} found at index {linear_result_found}" if linear_result_found != -1 else f"Linear Search (unsorted): Element {target_element_found} not found")

linear_result_not_found = linear_search(my_list_unsorted, target_element_not_found)
print(f"Linear Search (unsorted): Element {target_element_not_found} found at index {linear_result_not_found}" if linear_result_not_found != -1 else f"Linear Search (unsorted): Element {target_element_not_found} not found")

print("\nTesting Binary Search:")
binary_result_found = binary_search(my_list_sorted, target_element_found)
print(f"Binary Search (sorted): Element {target_element_found} found at index {binary_result_found}" if binary_result_found != -1 else f"Binary Search (sorted): Element {target_element_found} not found")

binary_result_not_found = binary_search(my_list_sorted, target_element_not_found)
print(f"Binary Search (sorted): Element {target_element_not_found} found at index {binary_result_not_found}" if binary_result_not_found != -1 else f"Binary Search (sorted): Element {target_element_not_found} not found")
```

explain when Binary Search is preferable.
Binary Search is preferable to Linear Search primarily when you are searching in a large, sorted dataset.

Here's why:

Time Complexity: Binary Search has a time complexity of O(log n), while Linear Search has a time complexity of O(n) in the worst and average cases. This means that as the size of the dataset (n) increases, the time it takes for Binary Search to find an element grows much slower than Linear Search. For large datasets, this difference in performance is significant.
Efficiency: Binary Search works by repeatedly dividing the search interval in half. This quickly eliminates a large portion of the data with each comparison, making it very efficient for finding elements in large sorted lists. Linear Search, on the other hand, checks each element one by one, which can be very time-consuming for large datasets.
However, Binary Search has a crucial requirement: the input array must be sorted. Use Binary Search when you have a large, sorted dataset and need to perform frequent searches.
Use Linear Search when you have a small dataset or an unsorted dataset and the cost of sorting outweighs the benefits of Binary Search for a single search.

# Task 4: Quick Sort and Merge Sort Comparison
•	Task: Implement Quick Sort and Merge Sort using recursion.
•	Instructions:

```python
def quick_sort(arr):
    """
    Sorts a list using the Quick Sort algorithm (recursive).

    Args:
        arr: The list to be sorted.

    Returns:
        A new list containing the sorted elements.

    Performance Notes:
        - Average Time Complexity: O(n log n)
        - Best Case Time Complexity: O(n log n)
        - Worst Case Time Complexity: O(n^2) (occurs with poor pivot selection, e.g., already sorted or reverse-sorted arrays)
        - Space Complexity: O(log n) on average due to recursion stack, O(n) in worst case.
    """
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        # Complete the recursive calls to fully sort the list
        return quick_sort(left) + middle + quick_sort(right)

def merge_sort(arr):
    """
    Sorts a list using the Merge Sort algorithm (recursive).

    Args:
        arr: The list to be sorted.

    Returns:
        A new list containing the sorted elements.

    Performance Notes:
        - Time Complexity: O(n log n) in all cases (best, average, and worst).
        - Space Complexity: O(n) due to the need for temporary arrays during merging.
        - It is a stable sort (maintains the relative order of equal elements).
    """
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Complete the recursive calls to sort the halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Implement the merge logic to combine the sorted halves
    merged_arr = []
    i = j = 0
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            merged_arr.append(left_half[i])
            i += 1
        else:
            merged_arr.append(right_half[j])
            j += 1
    while i < len(left_half):
        merged_arr.append(left_half[i])
        i += 1
    while j < len(right_half):
        merged_arr.append(right_half[j])
        j += 1

    return merged_arr

# Example Usage (You can test with sample data after completing the functions)
my_list = [3, 6, 8, 10, 1, 2, 1]
print("Quick Sort:", quick_sort(my_list.copy()))
print("Merge Sort:", merge_sort(my_list.copy()))
```

```python
import time
import copy

def measure_sort_time(sort_func, arr):
    """
    Measures the execution time of a given sorting function on a list.

    Args:
        sort_func: The sorting function to measure (e.g., quick_sort, merge_sort).
        arr: The list to be sorted.

    Returns:
        The elapsed time in seconds.
    """
    # Record the start time
    start_time = time.time()

    # Call the sorting function on a copy of the list
    sorted_arr = sort_func(copy.copy(arr))

    # Record the end time
    end_time = time.time()

    # Calculate and return the elapsed time
    return end_time - start_time

print("measure_sort_time function defined.")
```

```python
def find_duplicates_brute_force(arr):
    """
    Finds duplicate elements in a list using a brute-force approach with nested loops.

    Args:
        arr: The input list.

    Returns:
        A list of duplicate elements found in the input list.
        Note: This list may contain multiple occurrences of the same duplicate element.

    Time Complexity: O(n^2) - Due to the nested loops, where 'n' is the number of elements in the list.
    """
    duplicates = []
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] == arr[j] and arr[i] not in duplicates:
                duplicates.append(arr[i])
    return duplicates

# Example Usage
my_list = [1, 2, 3, 4, 2, 5, 6, 3, 7, 8, 8]
duplicate_items = find_duplicates_brute_force(my_list)
print("Duplicate elements (brute force):", duplicate_items)
```

```python
def find_duplicates_optimized(arr):
    """
    Finds duplicate elements in a list using an optimized approach with a set.

    Args:
        arr: The input list.

    Returns:
        A list of unique duplicate elements found in the input list.

    Time Complexity: O(n) - On average, where 'n' is the number of elements in the list.
                     Set insertion and lookup operations take O(1) time on average.
    Space Complexity: O(n) - In the worst case, where all elements are unique and stored in the set.
    """
    seen = set()
    duplicates = set()
    for item in arr:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)

# Example Usage
my_list = [1, 2, 3, 4, 2, 5, 6, 3, 7, 8, 8]
duplicate_items_optimized = find_duplicates_optimized(my_list)
print("Duplicate elements (optimized):", duplicate_items_optimized)
```

### Explanation of Optimization

The optimized approach uses a `set` to keep track of elements encountered so far. Here's how it achieves O(n) time complexity:

1.  **Single Pass:** The algorithm iterates through the list only once.
2.  **Set Lookups and Insertions:** For each element, it checks if the element is already in the `seen` set (`item in seen`). Set lookups and insertions have an average time complexity of O(1).
3.  **Adding Duplicates:** If an element is found in the `seen` set, it means it's a duplicate, and it's added to the `duplicates` set (also an O(1) average operation). If it's not in `seen`, it's added to the `seen` set.

In contrast, the brute-force method uses nested loops, resulting in O(n^2) comparisons in the worst case. The optimized approach avoids these nested comparisons by leveraging the efficient lookup capabilities of sets. The overall time complexity is dominated by the single pass through the list and the O(1) average time operations of the set, resulting in an average time complexity of O(n).

The space complexity is O(n) in the worst case because, if all elements are unique, the `seen` set will store all 'n' elements.
